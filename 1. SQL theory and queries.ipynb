# 10.Where

```python
# Operators in The WHERE Clause

=    >> Equal 
>    >> Greater than
<    >> Less than
>=   >> Greater than or equal
<=   >> Less than or equal
!=   >> not equal
```

```python
eg.

select * from emp where empid=1001;                    # = operator
select lname,empid,fname from emp where empid=1001;    # = operator
select * from emp where salary<40000;                  # < operator
select * from emp where salary>40000;                  # > operator
```

# 11.AND, OR and NOT Operators

```python
Eg.

select * from emp where empid=1001 or empid=1002;      # or operator
select * from emp where empid=1001 and empid=1002;     # and operator
select * from emp where fname='shital' and lname='kedar'; # and operator
select * from emp where not fname='shital'; # not operator
```

# 12. IN and NOT IN operator

```python
The IN operator allows you to specify multiple values in a WHERE clause.

The IN operator is a shorthand for multiple OR conditions.

syntax: SELECT column_name(s) FROM table_name WHERE column_name IN (value1, value2, ...);  # In sytnax
        SELECT column_name(s) FROM table_name WHERE column_name NOT IN (value1, value2, ...); # Not In syntax
        
eg.
select * from emp where empid in (1001,1002);          # in operator
select * from emp where empid not in (1001,1002);      # not in operator
```

# 13.BETWEEN and NOT BETWEEN Operator

```python
The BETWEEN operator selects values within a given range. The values can be numbers, text, or dates.

The BETWEEN operator is inclusive: begin and end values are included. 
    
syntax:
    SELECT column_name(s) FROM table_name WHERE column_name BETWEEN value1 AND value2; # between syntax
    
    SELECT column_name(s) FROM table_name WHERE column_name NOT BETWEEN value1 AND value2; # not between syntax
    
eg. select * from emp where salary between 30000 and 50000;        #  between(both values are included)
    select * from emp where salary not between 30000 and 50000;    # not between
```

# 14.LIKE Operator

```python
search for a pattern
The LIKE operator is used in a WHERE clause to search for a specified pattern in a column.

syntax: SELECT column1, column2, ... FROM table_name WHERE column LIKE pattern;

SELECT * FROM Customers WHERE CustomerName LIKE 'a%';
    
The percent sign (%) represents zero or more characters
The underscore sign (_) represents one character
```

# 15.NULL Values(IS NULL and IS NOT NULL operators)

```python
A field with a NULL value is a field with no value.
It is not possible to test for NULL values with comparison operators, such as =, <, or !=.
We will have to use the IS NULL and IS NOT NULL operators instead.

# IS NULL Syntax
select column_names from table_name where column_name IS NULL;

# IS NOT NULL Syntax
select column_names from table_name where column_name IS NOT NULL;
```

```python
eg.

SELECT CustomerName, ContactName, Address FROM Customers WHERE Address IS NULL; # Null 

SELECT CustomerName, ContactName, Address FROM Customers WHERE Address IS NOT NULL;  # NOT NULL

select * from emp where loc is null;
select * from emp where not loc is null;
select * from emp where loc is not null;
```

# 22.Limit clause

```python
It is use the access records from a table

# syntax-1: without where clause
SELECT column_name(s) FROM table_name LIMIT row_number;          # select top n rows
SELECT column_name(s) FROM table_name LIMIT offset,row_number; # select n rows after skipping some rows


# syntax-2: Limit Clause with Where clause
SELECT column_name(s) FROM table_name WHERE condition LIMIT number; # select top n rows
SELECT column_name(s) FROM table_name WHERE condition LIMIT (offset,row_number); 
                                                        # select n rows after skipping some rows                        
```

```python
eg.
select * from emp limit 2;     # Display top 2 rows
select * from emp limit 2,3;   # Skip first 2 rows and display next 3 rows
```

# 23.Order by

```python
order by - sort the data in ascending or descending order.   
The ORDER BY keyword sorts the records in ascending order by default. 
To sort the records in descending order, use the DESC keyword.

syntax: select column1, column2, ...from table_name order by column1, column2, ...; # Sort in ascending order
    
        select column1, column2, ...from table_name order by column1, column2, ... desc ; # Sort in descending order
        
        
eg. SELECT * FROM Customers ORDER BY Country;     # Sort single column in ascending order
    SELECT * FROM Customers ORDER BY Country desc; # Sort single column in descending order
        SELECT * FROM Customers ORDER BY Country, CustomerName;  # Sort two columns in ascending order
        SELECT * FROM Customers ORDER BY Country ASC, CustomerName DESC; 
                                   # Sort 1 column in ascending order and 1 column in descending order
```

# 24.Arithematic Expressions/Aggregation operation(SUM,AVG,MAX,MIN,COUNT)

```python
# 1. MIN(): This function returns the smallest value of the selected column.

syntax: SELECT MIN(column_name) FROM table_name WHERE condition;

eg. select min(price) from cars93;
```

```python
# 2. MAX(): This function returns the largest value of the selected column.

syntax: SELECT MAX(column_name) FROM table_name WHERE condition;
    
eg. select max(price) from cars93;
```

```python
# 3.COUNT: This function returns the number of rows that matches a specified criterion.

syntax: SELECT COUNT(column_name) FROM table_name WHERE condition;
    
eg. select count(price) from cars93;
```

```python
# 4. AVG: This function returns the average value of a numeric column.

syntax: SELECT AVG(column_name) FROM table_name WHERE condition;
    
eg. select avg(price) from cars93;
```

```python
# 5. SUM: This function returns the total sum of a numeric column. 

syntax: SELECT SUM(column_name) FROM table_name WHERE condition;
    
eg. select sum(price) from cars93;
```
